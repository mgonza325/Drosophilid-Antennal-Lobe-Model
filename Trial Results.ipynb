```python
%matplotlib inline
import matplotlib.pyplot as plt
import pytry
import numpy as np
import seaborn as sns
import pandas
```

```python
df = pandas.DataFrame(pytry.read('exp1'))
```

```python
weights = np.unique(df.w_ORN_Picky)

for w in weights:
    plt.figure(figsize=(12,3))
    fr = np.mean(df.query('w_ORN_Picky==%g'%w).responses_Picky)
    sns.heatmap(fr, cmap='Reds', vmin=0, vmax=200)
    plt.title('w=%g' % w)
    plt.show()

```
